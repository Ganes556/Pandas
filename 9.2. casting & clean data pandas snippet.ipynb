```python
import pandas as pd
import numpy as np
```

```python
people= {"first" : ["agung","NA","mandako","budi",np.nan,None],
        "last":["dwipa","Missing",np.nan,np.nan,"juliantari","tugus"],
        "email": ["agungdwipa@gmail.com","kuraoanas@gmail.com","NA","budi@mail.com",np.nan,None],
        "age": ["12","14",None,np.nan,None,"Missing"]}
```

```python
data= pd.DataFrame(people)
data.replace("NA",np.nan,inplace=True)
data.replace("Missing",np.nan,inplace=True)
```

```python
data
```

```python
data.dropna(how='all',subset=["last","email"],inplace=True)
```

```python
data
```

```python
data.isna()
```

```python
data.fillna(0,inplace=True)
```

```python
data["age"] = data["age"].astype(float) # cannot change type string
```
