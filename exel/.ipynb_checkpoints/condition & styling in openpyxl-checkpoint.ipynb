```python
from openpyxl import Workbook
from openpyxl import load_workbook
from openpyxl.styles import *
from openpyxl.formatting.rule import Rule
from openpyxl.styles.differential import DifferentialStyle
```

```python
wb = load_workbook("data/data2_excel.xlsx")
ws = wb.worksheets[0]
```

```python
## styling
for columns in ws.iter_rows(min_col=6,max_row=1):
    for column in columns:
#         print(column.value)
        column.font = Font(bold=True)
        column.border = Border(left=Side(style="thick"),
                              top=Side(style="thick"),
                              right=Side(style="thick"),
                              bottom=Side(style="thick"))
```

```python
# CONDITIONS
red_background = PatternFill(bgColor="FF0000")
diff_style = DifferentialStyle(fill=red_background) # seperti NamedStyle
rule = Rule(type="expression",dxf=diff_style)
rule.formula= ["$D2 > $E$2"]
ws.conditional_formatting.add("D2:D12",rule)
```

```python
wb.save(filename="data/data2_excel.xlsx")
```
