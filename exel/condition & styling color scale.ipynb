```python
from openpyxl.styles import *
from openpyxl.formatting.rule import ColorScaleRule
from openpyxl import Workbook
from openpyxl import load_workbook
```

```python
wb = load_workbook("data/data1.xlsx")
ws = wb.worksheets[0]
```

```python
## styling
kuning = "FFFF00"
for columns in ws.iter_rows(max_col=1,max_row=1):
    for column in columns:
        column.border = Border(left=Side(style="thick"),
                           top=Side(style="thick"),
                           right=Side(style="thick"),
                           bottom=Side(style="thick"))
        column.fill = PatternFill(start_color=kuning,fill_type="solid")
        
for columns in ws.iter_rows(max_col=1,min_row=2):
    for column in columns:
#         print(column.value)
        column.border = Border(left=Side(style="thin"),
                           top=Side(style="thin"),
                           right=Side(style="thin"),
                           bottom=Side(style="thin"))
```

```python
# formatting color scale rule
hijau = "008000"
merah = "FF0000"
color_scale_rule = ColorScaleRule(start_type= "formula=$D2 > $D$12",
                                   start_color=merah,
                                   end_type="formula=$D2 < $D$12",
                                   end_color=hijau)
ws.conditional_formatting.add("D2:D12",color_scale_rule)
```

```python
wb.save("data/color scale.xlsx")
```
