```python
country_grp["LanguageWorkedWith"].str.contains('Python').sum()
```

```python
country_response = public_df["Country"].value_counts()
```

```python
use_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains('Python').sum())
use_python
```

```python
python_df = pd.concat([country_response,use_python],axis="columns",sort=False)
python_df
```

```python
python_df.rename(columns={"Country":"Sum respons","LanguageWorkedWith":"Use python"},inplace=True)
```

```python
python_df
```

```python
python_df["Pct Use Python"] = (python_df["Use python"]/python_df["Sum respons"])*100 
python_df
```

```python
python_df.sort_values(by="Pct Use Python",ascending=False,inplace=True)
```
