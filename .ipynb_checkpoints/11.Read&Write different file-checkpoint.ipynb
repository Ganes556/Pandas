```python
from sqlalchemy import create_engine
import psycopg2
```

```python
# menggunakan sql file read dan writenya harus pakai postgresql
# contoh penggunaan
```

```python
#### WRITE DATA SQL ####
engine = create_engine("postgresql://dbuser:dbpass@localhost:5432/sample_db") # koneksi ke postgresql
indo_df.to_sql("sample_table",engine)# "name file" , connection
# jika data suda ada di postgresql bisa melakukan hal ini :
indo_df.to_sql("sample_table",engine,if_exist="replace") # datanya di replace
```

```python
sql_df = pd.read_sql("sample_table",engine,index_col="Respondent") # name file, connection
```

```python
sql_df.head()
```

```python
sql_df = pd.read_sql_query("SELECT * FROM sample_table",engine,index_col="Respondent") # bisa menjalankan query
sql_df.head() # run again
```
