```python
people= {"first" : ["agung","kurao","mandako"],
        "last":["dwipa","anas","anas"],
        "email": ["agungdwipa@gmail.com","kuraoanas@gmail.com","mandakoanas@gmail.com"]}
```

```python
import pandas as pd
data = pd.DataFrame(people)
```

```python
# updating columns
data.columns = ["first_name","last_name","email"]
```

```python
data.columns = [x.lower() for x in data.columns]
```

```python
data.columns = data.columns.str.replace(" ","_")
```

```python
data.rename(columns={"first_name":"first","last_name":"last"},inplace=True)
```

```python
data
```

```python
data.loc[2]=["agusss","tes","tes@gmail.com"]
```

```python
data.at[2,'last'] = "anos" # menggunakan at lebih cepat untuk single value
```

```python
filt = (data["last"] == "anas")
```

```python
data[filt]["last"] = "anus" # akan memberikan error jika ingin setting data
```

```python
data.loc[filt,"last"] = 'anos' # gunakan .loc untuk setting data
```

```python
data
```

```python
data["first"] = data["first"].str.lower()
```

```python
data
```

```python
data.apply(len)
```

```python
def upper_email(email):
    return email.upper()
```

```python
data["email"] = data["email"].apply(upper_email) # cara biasa
```

```python
data
```

```python
data["email"] = data["email"].apply(lambda x: x.lower()) # menggunakan lambda
```
