```python
people= {"first" : ["agung","kurao","mandako"],
        "last":["dwipa","anas","anas"],
        "email": ["agungdwipa@gmail.com","kuraoanas@gmail.com","mandakoanas@gmail.com"]}
```

```python
import pandas as pd
data = pd.DataFrame(people)
```

```python
data
```

```python
filt = (data["last"] == "anas") | (data["first"]=="kurao")
filt2 = (data["last"] == "anas") & (data["first"] == "mandako")
```

```python
data.loc[-filt,"email"] # '-' mengembalikan false value
```

```python
data.loc[filt2,"email"] # normal tanpa "-" dan menggunakan operator "&"
```

```python
filt_data = data["email"].str.contains("anas")
```

```python
data.loc[-filt_data,"email"]
```
